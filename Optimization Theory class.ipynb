```python
import numpy as np
import sympy as sp
from sympy import simplify, Matrix, hessian, Symbol, factor, solve,plot_implicit
from sympy.abc import x,y,z
from matplotlib.pyplot import show
sp.init_printing() 
```

```python
class Optimization:
    def __init__(self,f,var):
        self.f = f
        self.var = var
    
    def __repr__(self):
        return f'{self.f}'
    def __str__(self):
        return f'{self.f}'
    def __add__(self,other):
        return self.f + other.f
    def __sub__ (self,other):
        return self.f - other.f
    def __mul__ (self,other):
        return self.f * other.f
    def __rmul__ (self,val):
        return val * self.f
        
    def varf(self):
        Var_L = [Symbol(char) for char in self.var]
        return Var_L

    def vec(self):
        return Matrix(self.varf())
        
    def H(self):
        Hessianf = simplify(hessian(self.f,self.vec()))
        return Hessianf
    def Solve(self):
        return solve(self.f)
    
    def xHx(self,a_): #a_: grad(vec) = 0
        A = [i-j for i,j in zip(self.varf(),a_)]
        return factor(((1/2)*Matrix(np.array(A)).T@self.H()@Matrix(np.array(A)))[0])
    
    def plotf(self):
        return plot_implicit(self.f)
        
                
```

```python
function = x**2 -x*y +2*x +y**2 +2*y +2
f = Optimization(function,['x','y'])
```

```python
f.varf()
```

```python
f.vec()
```

```python
f.H()
```

```python
f.Solve()
```

```python
f.xHx([-2,-2])
```

```python
f.plotf()
plot_implicit(f.xHx([-2,-2])<1)
show()
```
